# 🎲 Prueba de Independencia — Prueba de Rachas

---

## Introducción

La **prueba de rachas** (o *runs test*) es una prueba no paramétrica que permite verificar si una secuencia de números aleatorios es **estadísticamente independiente**, es decir, si los valores no presentan patrones sistemáticos de subidas o bajadas.

### Hipótesis

$$H_0: \text{Los números de la muestra } \{X_i\} \text{ son independientes}$$

$$H_1: \text{Los números de la muestra } \{X_i\} \text{ no son independientes}$$

---

## Paso 1 — Construcción de la secuencia $S$

A partir de la muestra $\{X_1, X_2, \ldots, X_n\}$ se construye una secuencia $S$ de $n-1$ símbolos binarios de la siguiente forma:

$$
S_i = \begin{cases}
0 \;(\text{ó } {-}) & \text{si } X_i \leq X_{i-1} \\
1 \;(\text{ó } {+}) & \text{si } X_i > X_{i-1}
\end{cases}
\qquad i = 2, 3, \ldots, n
$$

La secuencia $S$ contiene $n - 1$ elementos.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from scipy.stats import norm

# -------------------------------------------------------
# Tabla 3.1 — ingresada por filas, leída por COLUMNAS
# -------------------------------------------------------
tabla = np.array([
    [0.78961, 0.05230, 0.10699, 0.55877, 0.14151],
    [0.76086, 0.12079, 0.27738, 0.65726, 0.79269],
    [0.80548, 0.82654, 0.29453, 0.20852, 0.42989],
    [0.58518, 0.98611, 0.34488, 0.34358, 0.11537],
    [0.89898, 0.57880, 0.67621, 0.05010, 0.00121],
    [0.28269, 0.73059, 0.70119, 0.18284, 0.49962],
    [0.38618, 0.76910, 0.68334, 0.55170, 0.10850],
    [0.79982, 0.45679, 0.21631, 0.87616, 0.55743],
    [0.58962, 0.33216, 0.03185, 0.61168, 0.09264],
    [0.69623, 0.17028, 0.05475, 0.91512, 0.76262],
    [0.29931, 0.30861, 0.83358, 0.51781, 0.03272],
    [0.57410, 0.26593, 0.85903, 0.43308, 0.35286],
    [0.24000, 0.65559, 0.38507, 0.90829, 0.94187],
    [0.93655, 0.88809, 0.81772, 0.36982, 0.19904],
    [0.54325, 0.62400, 0.09133, 0.41678, 0.33954],
    [0.58244, 0.85853, 0.88752, 0.33729, 0.15506],
    [0.23949, 0.53559, 0.33381, 0.49383, 0.75103],
    [0.19962, 0.65002, 0.74579, 0.79113, 0.63453],
    [0.19147, 0.40644, 0.08128, 0.73435, 0.22724],
    [0.22287, 0.07281, 0.64183, 0.44267, 0.72102]
])

# Leer por columnas (Fortran / column-major order)
numeros = tabla.flatten(order='F')
n = len(numeros)

print(f"Total de números: n = {n}")
print(f"Primeros 10 (columna 1): {numeros[:10]}")
```

```python
# --- Construcción de la secuencia S ---
S = np.where(numeros[1:] > numeros[:-1], 1, 0)   # 1 = sube (+), 0 = baja (-)

simbolos = np.where(S == 1, '+', '-')
print(f"Longitud de S: {len(S)} = n - 1 = {n} - 1\n")
print("Secuencia S:")
# Imprimir en grupos de 20 para mejor lectura
for i in range(0, len(S), 20):
    print('  ' + ' '.join(simbolos[i:i+20]))
```

---

## Paso 2 — Conteo de rachas observadas $C_0$

Una **racha** (*run*) es una sucesión maximal de símbolos iguales consecutivos dentro de $S$. Por ejemplo:

$$S = \{\underbrace{- -}_{\text{racha 1}} \underbrace{+}_{\text{2}} \underbrace{-}_{\text{3}} \underbrace{+ +}_{\text{4}} \cdots \}$$

$C_0$ es el número total de rachas contadas en $S$.

```python
# --- Conteo de rachas ---
# Una nueva racha comienza cada vez que S_i != S_{i-1}
C0 = 1 + np.sum(S[1:] != S[:-1])

print(f"Número de rachas observadas: C₀ = {C0}")
```

---

## Paso 3 — Valor esperado, varianza y estadístico $Z_0$

Bajo $H_0$, el número de rachas $C_0$ sigue aproximadamente una distribución normal con:

$$\mu_{C_0} = \frac{2n - 1}{3}$$

$$\sigma^2_{C_0} = \frac{16n - 29}{90}$$

El estadístico de prueba estandarizado es:

$$Z_0 = \frac{C_0 - \mu_{C_0}}{\sigma_{C_0}}$$

donde $Z_0 \sim \mathcal{N}(0,\,1)$ bajo $H_0$.

```python
# --- Cálculo de estadísticos ---
mu_C0     = (2 * n - 1) / 3
sigma2_C0 = (16 * n - 29) / 90
sigma_C0  = np.sqrt(sigma2_C0)
Z0        = (C0 - mu_C0) / sigma_C0

print(f"n             = {n}")
print(f"C₀            = {C0}")
print(f"μ_C₀          = (2×{n}−1)/3          = {mu_C0:.4f}")
print(f"σ²_C₀         = (16×{n}−29)/90        = {sigma2_C0:.4f}")
print(f"σ_C₀          = √{sigma2_C0:.4f}              = {sigma_C0:.4f}")
print(f"Z₀            = ({C0} − {mu_C0:.4f}) / {sigma_C0:.4f} = {Z0:.4f}")
```

---

## Paso 4 — Intervalo de aceptación y decisión

Para un nivel de significancia $\alpha$, la **región de aceptación** bilateral es:

$$-Z_{\alpha/2} \leq Z_0 \leq Z_{\alpha/2}$$

**Regla de decisión:**

$$
\begin{cases}
\text{No rechazar } H_0 & \text{si } |Z_0| \leq Z_{\alpha/2} \\
\text{Rechazar } H_0    & \text{si } |Z_0| > Z_{\alpha/2}
\end{cases}
$$

```python
# --- Decisión ---
alpha    = 0.05
Z_alpha2 = norm.ppf(1 - alpha / 2)

print(f"Nivel de significancia α = {alpha}")
print(f"Valor crítico Z_{{α/2}}   = ±{Z_alpha2:.4f}")
print(f"Intervalo de aceptación : [{-Z_alpha2:.4f}, {Z_alpha2:.4f}]")
print(f"Z₀                      = {Z0:.4f}")
print()

if abs(Z0) <= Z_alpha2:
    print("✔  CONCLUSIÓN: No se rechaza H₀.")
    print(f"   Los números SON independientes (α = {alpha}).")
else:
    print("✘  CONCLUSIÓN: Se rechaza H₀.")
    print(f"   Los números NO son independientes (α = {alpha}).")
```
